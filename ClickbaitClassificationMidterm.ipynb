## Preparing the data

Here we will Tokenise, Remove stop words, punctuation and peform Lemmatizeation on the text.


### Tokenize the headlines ("Text")

```python


def tokenization(text):
    if isinstance(text, str):
        lst = text.split()
        return lst
    else:
        return []
x_train = [tokenization(t) for t in x_train]
x_test = [tokenization(t) for t in x_test]
```

### Remove Stop Words

Stop words are words that are commonly used in text but do not contain much meaning. They are usually removed from text data to avoid adding unnecessary noise to the data and to improve the performance of natural language processing (NLP) models. Some examples of stop words include articles (a, an, the), conjunctions (and, or, but), and prepositions (in, on, at). Removing stop words can help to focus the analysis on the most meaningful words in the text.

```python

def remove_stopwords(lst):
    stop=stopwords.words('english')
    new_lst=[]
    for i in lst:
        if i not in stop:
            new_lst.append(i)
    return new_lst

x_train = [remove_stopwords(x) for x in x_train]
x_test = [remove_stopwords(x) for x in x_test]
```

### Removing Punctuation

```python
def remove_punctuations(lst):
    new_lst=[]
    for i in lst:
        for j in s.punctuation:
            i=i.replace(j,'')
        new_lst.append(i)
    return new_lst
x_train=[remove_punctuations(x) for x in x_train]
x_test=[remove_punctuations(x) for x in x_test]
```

### Lemmatization

Lemmatization is the process of converting a word to its base form. The base form of a word is known as the lemma. For example, the lemma of the word "jumps" is "jump".

Lemmatization is useful because it helps to reduce the dimensionality of the data by reducing the number of words in the vocabulary. This can make it easier to analyze and process the text data. Additionally, lemmatization can help to improve the performance of some natural language processing tasks because it can reduce the impact of inflectional forms and spelling variations on the analysis.

```python
lemmatizer=nltk.stem.WordNetLemmatizer()
def lemmatzation(lst):
    new_lst=[]
    for i in lst:
        i=lemmatizer.lemmatize(i)
        new_lst.append(i)
    return new_lst
x_train = [lemmatzation(lst) for lst in x_train]
x_test = [lemmatzation(lst) for lst in x_test]
```

## BaseLine Peformance With A Dummy Classifier

A dummy classifier is a simple classifier that is used as a baseline model for comparison purposes. It is used to determine the minimum performance that can be achieved by a classifier, and to establish a baseline against which more complex models can be compared.

There are several reasons why we might use a dummy classifier:

To establish a baseline: By comparing the performance of a dummy classifier to a more complex model, we can determine how much improvement the complex model provides over the baseline.

To check the validity of the evaluation metric: If the dummy classifier performs significantly better than the complex model, it may indicate that the evaluation metric is not a good measure of performance for the task at hand.

To check the assumption of class balance: A dummy classifier that always predicts the majority class can be used to verify that the class balance of the dataset is reasonable, and to identify any issues with imbalanced classes.

To debug the model: If the dummy classifier performs significantly better than the complex model, it may indicate that there is an issue with the complex model, such as overfitting or a bug in the implementation.

```python
# Initialize the dummy classifier
dummy_classifier = DummyClassifier(strategy='most_frequent')

# Fit the classifier on the training data
dummy_classifier.fit(x_train, y_train)

# Evaluate the classifier on the test data
accuracy = dummy_classifier.score(x_test, y_test)
print(f'Test accuracy: {accuracy:.2f}')
```

### Classification report for the Dummy Classifier

```python
# Generate prediction on the test set
y_pred = dummy_classifier.predict(x_test)

# Print the evaluation report
print(classification_report(y_test, y_pred))
```

Precision is the proportion of true positive predictions made by the classifier out of all positive predictions. It is a measure of the accuracy of the classifier when it predicts the positive class.

Recall is the proportion of true positive predictions made by the classifier out of all actual positive instances. It is a measure of the classifier's ability to detect all positive instances.

The f1-score is the harmonic mean of precision and recall. It is a single metric that combines both precision and recall, and is a useful measure when you want to balance precision and recall.

Support is the number of instances of each class in the dataset. It is useful for understanding the class balance in the dataset and how well the classifier performs on each class.

In this particular evaluation,the classifier has poor precision and recall, as both metrics are close to 0 for class 0 and close to 1 for class 1. The overall f1-score and accuracy are also poor, at 0.33 and 0.50 respectively. This suggests that the classifier is not performing well on this task.

## LSTM Model

Long Short-Term Memory (LSTM) networks are a type of recurrent neural network (RNN) that are particularly well-suited to learning long-term dependencies in sequential data. They are commonly used in natural language processing tasks such as language translation and text classification, including clickbait detection.

Overall, LSTMs are a powerful tool for clickbait detection because they can handle the long-term dependencies, variable-length sequences, and contextual information that are common in clickbait headlines.

```python
 #Initialize the Tokenizer
tokenizer = Tokenizer(num_words=5000)

# Fit the Tokenizer on the training data
tokenizer.fit_on_texts(x_train)

# Convert the text to sequences
text_train_sequences = tokenizer.texts_to_sequences(x_train)
text_test_sequences = tokenizer.texts_to_sequences(x_test)
# Pad the sequences to the same length
text_train_padded = pad_sequences(text_train_sequences, maxlen=300)
text_test_padded = pad_sequences(text_test_sequences, maxlen=300)

# Initialize the model
model = Sequential()
# Add the embedding layer
model.add(Embedding(5000, 128, input_length=300))
# Add the LSTM layer
model.add(LSTM(128))
# Add the output layer
model.add(Dense(1, activation='sigmoid'))

```

```python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Train the model
history = model.fit(text_train_padded, y_train, epochs=5, validation_data=(text_test_padded, y_test))
```

```python
model.save('model')
```

### Plot of Training and validation accuracy for LSTM Model

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
x = range(1, len(acc) + 1)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(x, acc, 'b', label='Training acc')
plt.plot(x, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(x, loss, 'b', label='Training loss')
plt.plot(x, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

### Classification report for LSTM Model

```python

# Use the model to make predictions on the test data

x_test_array = np.asarray(x_test)


preds = model.predict(text_test_padded)


preds = np.round(preds).astype(int)

# Get the classification report
report = classification_report(y_test, preds)

print(report)
```

Based on the output, the LSTM model is performing well on the clickbait detection dataset. The model was trained for 5 epochs, and the training and validation accuracy improved consistently over the course of training. The final evaluation on the test set shows that the model has a high precision, recall, and f1-score for both classes, and an overall accuracy of 0.98. This indicates that the model is able to accurately classify clickbait and non-clickbait headlines.

It is worth noting that the validation loss and accuracy plateaued after the third epoch, while the training loss and accuracy continued to improve. This suggests that the model may be overfitting to the training data after the third epoch. To address this issue, it would be worth using techniques such as early stopping or dropout to prevent overfitting in future iterations of this project.

## Naive Bayes Model

Naive Bayes is a simple probabilistic classifier that is based on the Bayes theorem and the assumption of independence between features. It is commonly used in natural language processing tasks such as text classification, including clickbait detection.

There are several reasons why Naive Bayes might be used for clickbait detection:

Simplicity: Naive Bayes is a fast and easy-to-implement classifier that requires minimal preprocessing of the data.

Robustness: Naive Bayes is relatively robust to the presence of irrelevant features and is not sensitive to the scaling of the data.

Scalability: Naive Bayes scales well to large datasets, making it suitable for use with large collections of text data.

```python
nb = MultinomialNB(alpha=1.0)

# Now you can fit the Naive Bayes model on `x_train_array` and `y_train`
nb.fit(text_train_padded, y_train)

# Predict the labels of the test data
predictions = nb.predict(text_test_padded)
# Calculate the accuracy of the model
accuracy = nb.score(text_test_padded, y_test)
print(accuracy)
```
